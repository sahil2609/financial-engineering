```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
ce = pd.read_csv('INFY_CE.csv')
pe = pd.read_csv('INFY_PE.csv')
```

```python
ce["Date"] = pd.to_datetime(ce["Date"]).dt.strftime("%d-%m-%Y")
ce["Expiry"] = pd.to_datetime(ce["Expiry"]).dt.strftime("%d-%m-%Y")
ce = ce.loc[ce['Date'] <= '2023-02-28']
ce = ce.sort_values(by=['Date', 'Expiry'],ascending=False)
ce.head()
```

```python
pe["Date"] = pd.to_datetime(pe["Date"]).dt.strftime("%d-%m-%Y")
pe["Expiry"] = pd.to_datetime(pe["Expiry"]).dt.strftime("%d-%m-%Y")
pe = pe.loc[pe['Date'] <= '2023-02-28']
pe = pe.sort_values(by=['Date', 'Expiry'],ascending=False)
pe.head(5)
```

```python
arr = []
for col in ce.columns:
    if(str(col) != 'Date' and str(col)!= 'Symbol' and str(col)!='Expiry' and str(col)!='Strike Price'):
        if(str(col) == "Close_Call"):
            arr.append("Close_CE")
        else:
            arr.append(str(col) + "_CE")
    else:
        arr.append(str(col))
ce.columns = arr
ce.columns
```

```python
arr = []
for col in pe.columns:
    if(str(col) != 'Date' and str(col)!= 'Symbol' and str(col)!='Expiry' and str(col)!='Strike Price'):
        if(str(col) == "Close_Put"):
            arr.append("Close_PE")
        else:
            arr.append(str(col) + "_PE")
        
    else:
        arr.append(str(col))
pe.columns = arr
pe.columns
```

```python
ce = ce[['Symbol','Date','Expiry', 'Strike Price', 'Close_CE']]
ce.head()
```

```python
pe = pe[['Symbol','Date','Expiry', 'Strike Price', 'Close_PE']]
pe.head()
```

```python
final = ce.merge(pe[pe.columns])
final.head()
```

```python
final.to_csv('INFY.NS.csv')
```
